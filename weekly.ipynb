```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import datetime
from pprint import pprint
import random as rnd

# matplotlib setup
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = (15, 15)      # setting default size of plots
```

```python
df = pd.read_excel("data.xlsx")
print("Loaded {} rows and {} columns.".format(*df.shape))
```

```python
times = np.array(df.columns[1:])
sites = np.array(df["Sites"])

days = np.unique([time.day for time in times])

daily_sums = dict()

for day in days:
    daily_sum = dict()
    columns = [column for column in df.columns[1:] if column.day==day]
    
    for index, site in enumerate(sites):
        site_sum = 0

        for column in columns:
            site_sum += df[column][index]
        
        daily_sum[site] = site_sum
        
    daily_sums[datetime.date(2016, 1, day)] = daily_sum
```

```python
plottable = np.zeros((len(sites), len(days)))
for day_index, day in enumerate(daily_sums.keys()):
    for site_index, site in enumerate(daily_sums[day].keys()):
        plottable[site_index][day_index] = daily_sums[day][site]
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(
                   z=plottable,
                   x=days,
                   y=sites,
                   colorscale='delta'))
fig.show()
```
